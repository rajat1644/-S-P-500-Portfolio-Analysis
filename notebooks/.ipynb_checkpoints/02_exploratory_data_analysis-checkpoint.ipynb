```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
# loading cleaned data
df=pd.read_csv('./data/processed/cleaned_stocks.csv')
df['date'] = pd.to_datetime(df['date'])
print("Data timeframe:", df['date'].min(), "to", df['date'].max())
print("No. Of Companies", df['symbol'].nunique())
df.describe()
```

```python
# visualizing stocks' closing price trends over period of time
```

```python
selected_stocks = ['AAPL','MSFT','GOOGL']
for symbol in selected_stocks:
    stock_df = df[df['symbol']==symbol]
    plt.figure(figsize=(12,6))
    plt.plot(stock_df['date'], stock_df['close'], label=symbol)
    plt.title(f'{symbol} Stock Price Over Time')    
    plt.xlabel('Date')
    plt.ylabel('Close Price')
    plt.grid(False)
    plt.legend()
    plt.savefig('../reports/visuals/eda_stock_price.png')
    plt.show()

```

```python
# Visualizing Daily Returns
```

```python
df_sorted = df.sort_values(by=['symbol', 'date'])
df_sorted['daily_return'] = df_sorted.groupby('symbol')['close'].pct_change()
plt.figure(figsize=(14, 10))
sns.boxplot(data=df_sorted[df_sorted['symbol'].isin(selected_stocks)],
            x='symbol', y='daily_return')
plt.title('Daily Return Distribution')
plt.savefig('../reports/visuals/daily_return_distrvutions.png')
plt.show()
```

```python
returns_pivot=df_sorted.pivot(index='date', columns='symbol', values='daily_return')
returns_correlation = returns_pivot.corr()
# Heatmap
plt.figure(figsize=(10,8))
sns.heatmap(returns_correlation, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation of Daily Returns Betweeen Stocks')
plt.savefig('../reports/visuals/correlation_selected_stocks.png')
plt.show()
```

```python
### ✅ Summary:
# - Visualized stock trends (close prices over time).
# - Computed daily returns and analyzed their distribution.
# - Identified correlations between Apple, Microsoft, Google.
```
