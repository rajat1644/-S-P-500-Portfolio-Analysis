```python
# Build a supervised ML model (e.g., Linear Regression or Random Forest) 
# to predict future closing prices of ORCL stock, using features like 
# lag values, moving averages, etc.
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

```python
orcl_df = pd.read_csv('../individual_stocks_5yr/individual_stocks_5yr/ORCL_data.csv')
orcl_df.head()
```

```python
orcl_df.columns = orcl_df.columns.str.lower().str.strip()
orcl_df.rename(columns={'name': 'symbol'}, inplace=True)
orcl_df.isnull().sum()
orcl_df.info()
orcl_df.dropna(inplace=True)
orcl_df.drop_duplicates(inplace=True)
orcl_df.head()
```

```python
orcl_df.to_csv('./data/processed/orcl_clean_data.csv')
```

```python
orcl_df = orcl_df[orcl_df['symbol'] == 'ORCL'].copy().sort_values('date')
orcl_df.set_index('date', inplace=True)
orcl_df.head()
```

```python
# adding time-series
orcl_df['close_lag1'] = orcl_df['close'].shift(1)
orcl_df['close_lag2'] = orcl_df['close'].shift(2)
orcl_df['rolling_mean_7'] = orcl_df['close'].rolling(7).mean()
orcl_df['rolling_std_7'] = orcl_df['close'].rolling(7).std()
orcl_df.dropna(inplace=True)
```

```python
# splitting features and target
X = orcl_df[['close_lag1', 'close_lag2', 'rolling_mean_7', 'rolling_std_7']]
y = orcl_df['close']

X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=False, test_size=0.2)
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
rf = RandomForestRegressor(n_estimators = 100, random_state=42)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)

print("Linear Regression R2:\n", r2_score(y_test, y_pred))
print("Linear Regression RMSE:\n", np.sqrt(mean_squared_error(y_test, y_pred)))
print("Random Forest R2:\n", r2_score(y_test, y_pred_rf))
print("Random Forest RMSE:\n", np.sqrt(mean_squared_error(y_test, y_pred_rf)))
```

```python
# plot the predictions:
plt.figure(figsize=(10,5))
plt.plot(y_test.index, y_test.values, label='Actual')
plt.plot(y_test.index, y_pred_rf, label='Random Forest Predictions', linestyle='--')
plt.title('ORCL Stock Price Prediction (Random Forest)')
plt.legend()
plt.grid()
plt.savefig('../reports/visuals/orcl_prediction_random_forest.png')
plt.show()
```
