```python
import pandas as pd
import numpy as np
import os
```

```python
df = pd.read_csv(r"D:\S&P500-Portfolio-Analysis\all_stocks_5yr.csv")
df.head()
```

```python
df.columns = df.columns.str.lower().str.strip()
df.rename(columns={'name': 'symbol'}, inplace=True)
df.info()
df.isnull().sum()
df.head()
```

```python
df.dropna(inplace=True)
df.drop_duplicates(inplace=True)
```

```python
# only focus on Apple, microsoft google amazon and oracle
```

```python
tickers=['AAPL','MSFT','GOOGL','AMZN','ORCL']
df_clean=df[df['symbol'].isin(tickers)]
df_clean.info()
df_clean.head(10)
```

```python
# saving cleaned data file
```

```python
os.makedirs('data/processed',exist_ok=True)
```

```python
df_clean.to_csv('data/processed/cleaned_stocks.csv',index=False)
```
